```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 2000
ln = [[] for i in range(5)]
x = []
for i in range(1,n):
    k = i/100
    x.append(k)
    ln[0].append(np.log(k))
    ln[1].append(k)
  
plt.plot(x, ln[1], label='N')
plt.plot(x, ln[0], label='Log(N)')
plt.axhline(y=0, color='grey', linestyle='--')
plt.axvline(x=1, color='grey', linestyle='--')
plt.axvline(x=0, color='grey', linestyle='--')
plt.legend()
```

```python
# https://stackoverflow.com/questions/33214852/is-anything-less-than-n-is-log-n
n = 100
ln = [[] for i in range(5)]
x = []
for i in range(1,n):
    x.append(i)
    ln[0].append(i**2)
    ln[1].append(i*np.log(i))
    ln[2].append(i)
    ln[3].append(np.log(i))
    ln[4].append(np.log(np.log(i)))
    
plt.plot(x, ln[0], label='N^2')
plt.plot(x, ln[1], label='N*Log(N)')
plt.plot(x, ln[2], label='N')
plt.plot(x, ln[3], label='Log(N)')
plt.plot(x, ln[4], label='Log(Log(N))')
plt.ylim([0,100])
plt.legend()
```

```python
n = 500000
ln = [[] for i in range(5)]
x = []
for i in range(10,n,250):
    x.append(i)
    ln[0].append(np.sqrt(np.log(1+i)/i))
    ln[1].append(np.sqrt(np.log(i)/i))
    ln[2].append(np.sqrt(np.log(np.log(i))/i))
    ln[3].append(np.sqrt(1/i))

plt.figure(figsize=(20, 10))
plt.plot(x, ln[0], label='sqrt(np.log(1+n)/n)')
plt.plot(x, ln[1], '--', label='sqrt(np.log(n)/n)')
plt.plot(x, ln[2], '--', label='sqrt(np.log(np.log(n))/n)')
plt.plot(x, ln[3], '--', label='sqrt(1/n)')
plt.ylim([0,0.1])
plt.legend()
```
